### Introducción

El arroz es uno de los cultivos más importantes a nivel global, siendo la base alimentaria de más de la mitad de la población mundial. Entre las numerosas variedades existentes (más de 10,000), el Osmancik (originario de Turquía) y el Cammeo (producido en Italia) destacan por sus características únicas en textura, composición nutricional y usos culinarios. Sin embargo, la creciente demanda y los estándares de calidad en la industria alimentaria han generado la necesidad de desarrollar métodos eficientes para clasificar, autenticar y optimizar estos tipos de arroz, evitando adulteraciones y garantizando su correcta comercialización.

En este contexto, el análisis de datos y el machine learning emergen como herramientas clave para diferenciar automáticamente estas variedades a partir de sus atributos morfológicos (como longitud, área y perímetro del grano) y químicos (contenido de proteínas, humedad y cenizas). Estudios previos, como el dataset analizado en IJISAE (que incluye 3,810 muestras de Osmancik y 2,876 de Cammeo), demuestran que estas variables presentan patrones distinguibles, lo que abre la puerta a la aplicación de algoritmos de clasificación supervisada.

Este proyecto busca validar si es posible identificar con alta precisión estas variedades mediante modelos predictivos, lo que tendría aplicaciones directas en:

* Control de calidad en molinos y distribuidores de arroz.

* Detección de fraudes en productos etiquetados como "puros".

* Optimización agrícola, correlacionando características del grano con condiciones de cultivo.

# **Información General del Arroz**

El arroz es la semilla de la planta *Oryza sativa* (arroz asiático o simplemente arroz) o de *Oryza glaberrima* (arroz africano).

Existen más de 10,000 variedades de arroz y es el segundo cereal más producido en el mundo (741,5 millones de toneladas). A la vez que se podría decir que el arroz es el cereal más importante en la alimentación humana y que contribuye de forma muy efectiva al aporte calórico de la dieta humana actual.
Se cree que el arroz (Oryza sativa) se domesticó hace 8.000-10.000 años en la región del valle del Yangtsé (China) y en el subcontinente indio (Oryza indica).

### Importancia del arroz:
* **Alimentación básica:** Más de 3.500 millones de personas dependen del arroz como principal fuente de energía, especialmente en Asia, América Latina y África.

* **Valor nutricional:** Proporciona carbohidratos complejos, proteínas (aunque incompletas, se complementan con legumbres), vitaminas (B1, B3, ácido fólico) y minerales (hierro, magnesio).

* **Cultivo clave para la economía:** Es el sustento de millones de agricultores y un producto fundamental en el comercio internacional.

* **Versatilidad culinaria:** Se usa en platos principales, postres, bebidas (como el sake japonés) y hasta en la industria (harina, aceite de salvado).

### Top 5 Productores (2023)
1. **China** (≈ 210 millones de toneladas)
2. **India** (≈ 180 millones de toneladas)
3. **Indonesia** (≈ 55 millones de toneladas)
4. **Bangladés** (≈ 38 millones de toneladas)
5. **Vietnam** (≈ 27 millones de toneladas)


### Clasificación:
1. **Grano largo**: Ligero y esponjoso, ideal para platos como *pilafs* y *ensaladas*.
2. **Grano medio**: Más pegajoso, perfecto para *paellas* y *guisos cremosos*.
3. **Grano corto**: Su textura es suave y almidonada, ideal para *risottos* y *sushi*.

El arroz también se clasifica según su procesamiento:
- **Arroz integral**: Conserva el salvado y el germen, ofreciendo más nutrientes.
- **Arroz blanco**: Se elimina el salvado y el germen, dejando solo el endospermo.


## Sobre las variedades Cammeo y Osmancik
- **Cammeo**: Introducida en 2014, esta variedad de arroz de grano largo es conocida por su textura suave y su capacidad de absorción, lo que la hace ideal para platos como risottos. Tiene un aspecto vidrioso y opaco.
- **Osmancik**: Cultivada ampliamente en Turquía desde 1997, esta variedad también es de grano largo y se utiliza comúnmente en la cocina mediterránea. Es reconocida por su rendimiento y calidad en la producción agrícola.

Ambas variedades son populares por su adaptabilidad y características físicas únicas, lo que las hace ideales para estudios de clasificación y análisis de calidad.


### **Sobre el experimento**
El experimento con las variedades Cammeo y Osmancik se centra en la clasificación de los granos utilizando métodos de visión por computadora y aprendizaje automático. Aquí están los detalles clave:
- **Objetivo**: Identificar y clasificar las variedades de arroz basándose en características morfológicas.
- **Dataset**: Incluye imágenes de 3810 granos de arroz procesadas para extraer 7 características morfológicas:
  - Área: Número de píxeles dentro de los límites del grano.
  - Perímetro: Circunferencia calculada alrededor del grano.
  - Longitud del eje mayor: Línea más larga que se puede dibujar en el grano.
  - Longitud del eje menor: Línea más corta que se puede dibujar en el grano.
  - Excentricidad: Medida de qué tan redonda es la elipse que tiene los mismos momentos que el grano.
  - Área convexa: Número de píxeles del caparazón convexo más pequeño que rodea el grano.
  - Extensión: Relación entre el área del grano y los píxeles de la caja delimitadora.
- **Métodos**: Se aplicaron algoritmos de aprendizaje automático como regresión logística y Random Forest para clasificar los granos con alta precisión.

### **Importancia del estudio**
Este experimento no solo ayuda a mejorar la clasificación de variedades de arroz, sino que también tiene aplicaciones en la agricultura de precisión, permitiendo a los agricultores optimizar la producción y calidad del arroz.

### Entre sus aplicaciones prácticas estan:
- **Control de calidad:** Empresas pueden usar el modelo para verificar autenticidad en importaciones.

- **Mejoramiento genético:** Identificar qué rasgos morfológicos/químicos correlacionan con mayor rendimiento.

- **Blockchain alimentario:** Integrar predicciones en sistemas de trazabilidad.

# Code:


```python
import os
import joblib
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import (
    LabelEncoder,
    StandardScaler,
    PowerTransformer,
)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import(
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
    f1_score,
    accuracy_score,
    RocCurveDisplay,
)
from sklearn.inspection import(
    permutation_importance
)
```

### **1. Data loading**
**Objective**: Obtain the data from source and get a first glimpse of their properties and presentation

```python
df_raw = pd.read_excel('/workspaces/final_project_rice/data/Rice_Cammeo_Osmancik.xlsx')
df_raw.info()
```

```python
df_raw.sample(10, random_state=10)
```

### **2. Data preprocessing**
**Objectives**: Perform the data cleaning, data transformation and data reduction steps to avoid data mistmatching, noisy data or data not wrangled

```python
df_baking = df_raw.copy()
df_baking.columns = df_baking.columns.str.lower()
df_baking = df_baking[['area','perimeter','major_axis_length','minor_axis_length','eccentricity','convex_area','extent','class']]
df_baking['class'] = df_baking['class'].astype('category')
df = df_baking.copy()
df.info()
```

### **3. Exploratory Data Analysis**
**Objective**: Summarize the main characteristics of the dataset using descriptive statistics and data visualization methods

```python
df_train, df_test = train_test_split(
    df,
    test_size=0.2,
    stratify=df['class'],
    random_state=2025
)
df_train, df_val = train_test_split(
    df_train,
    test_size=0.15,
    stratify=df_train['class'],
    random_state=2025
)
df_train.reset_index(drop=True, inplace=True)
df_val.reset_index(drop=True, inplace=True)
df_test.reset_index(drop=True, inplace=True)
df_train.shape, df_val.shape, df_test.shape
```

```python
display(df_train.describe(include="number").T)
display(df_train.describe(include="category").T)
```

```python
sns.countplot(x="class", data=df_train)
plt.title("Distribution of Rice Classes")
plt.show()
```

```python
df_train.hist(figsize=(10,10))
plt.show()
```

```python
sns.pairplot(df_train, hue="class", palette="husl")
plt.show()
```

```python
#identificar outlier con boxplot
sns.boxplot(x="class", y="area", data=df_train)
plt.show()
```

```python
#heatmap para identificar la relación entre variables
numerical_df = df_train.select_dtypes(include=np.number)
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

### **4. Machine learning**
**Objective**: Create a model that learns from data to make predictions and generalize to unseen data, and thus perform tasks without explicit instructions

```python
# Random Forest Classifier


# Separar características (X) y etiquetas (y)
X_train = df_train.drop(columns=["class"])
y_train = df_train["class"]
X_val = df_val.drop(columns=["class"])
y_val = df_val["class"]
X_test = df_test.drop(columns=["class"])
y_test = df_test["class"]

# Codificar las etiquetas
label_encoder = LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
y_val_enc = label_encoder.transform(y_val)
y_test_enc = label_encoder.transform(y_test)

# Definir el Pipeline
clf_rforest_p = Pipeline([
    ("powertransform", PowerTransformer()),
    ("rforest", RandomForestClassifier(random_state=2025, class_weight="balanced"))
])

# Optimizar hiperparámetros con GridSearchCV
clf_rforest_cv = GridSearchCV(
    clf_rforest_p,
    param_grid = {
    "rforest__max_depth": [ 5, 15],
    "rforest__n_estimators": [30, 50],
    "rforest__min_samples_split": [2, 5, 10],
    "rforest__min_samples_leaf": [2, 3, 4],
    "rforest__bootstrap": [True, False]
    },
    cv=5,
    scoring="f1",
    n_jobs=2
)

# Entrenar el modelo
clf_rforest_cv.fit(X_train, y_train_enc)

# Imprimir mejores parámetros
print("Mejores parámetros:", clf_rforest_cv.best_params_)

# Generar predicciones
y_pred = clf_rforest_cv.predict(X_test)

# Calcular métricas de rendimiento
accuracy = accuracy_score(y_test_enc, y_pred)
f1 = f1_score(y_test_enc, y_pred, average='weighted')

# Mostrar resultados
print(f"Random Forest Accuracy: {accuracy:.2f}")
print(f"F1 Score: {f1:.2f}")
print("\n")

# Clasification Report
print(classification_report(y_test_enc, y_pred, target_names=label_encoder.classes_))
```

```python
# Confusion Matrix Random Forest

cm = confusion_matrix(y_test_enc, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```

```python
# ROC Curve Random Forest
 
RocCurveDisplay.from_estimator(clf_rforest_cv.best_estimator_, X_test, y_test_enc)
plt.title("ROC Curve - Random Forest")
plt.show()
```

```python
# Permutation Importance Random Forest

perm_importance = permutation_importance(
    clf_rforest_cv.best_estimator_, X_test, y_test_enc, scoring="accuracy", n_repeats=10, random_state=2025
)

# Extraer las características y su importancia
importances = perm_importance.importances_mean
features = X_test.columns

# Ordenar los resultados
sorted_idx = np.argsort(importances)[::-1]

# Graficar
plt.figure(figsize=(10,6))
plt.barh(features[sorted_idx], importances[sorted_idx], color="blue")
plt.xlabel("Importancia")
plt.ylabel("Características")
plt.title("Permutation Importance - Random Forest")
plt.show()
```

```python
# Logiscitic Regression Model

log_reg_model = LogisticRegression(max_iter=1000, random_state=2025)
log_reg_model.fit(X_train, y_train_enc)
y_pred_log_reg = log_reg_model.predict(X_test)
accuracy_log_reg = accuracy_score(y_test_enc, y_pred_log_reg)
f1_log_reg = f1_score(y_test_enc, y_pred_log_reg, average='weighted')

# Mostrar resultados
print(f"Logistic Regression Accuracy: {accuracy_log_reg:.2f}")
print(f"F1 Score: {f1_log_reg:.2f}")
print("\n")
# Classification Report
print(classification_report(y_test_enc, y_pred_log_reg, target_names=label_encoder.classes_))
```

```python
# Confusion Matrix Logistic Regression

cm_log_reg = confusion_matrix(y_test_enc, y_pred_log_reg)
disp_log_reg = ConfusionMatrixDisplay(confusion_matrix=cm_log_reg, display_labels=label_encoder.classes_)
disp_log_reg.plot(cmap='Blues')
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

```python
# ROC Curve Logistic Regression
RocCurveDisplay.from_estimator(log_reg_model, X_test, y_test_enc)
plt.title("ROC Curve - Logistic Regression")
plt.show()
```

```python
#Permutation Importance Logistic Regression

# Calcular la importancia de las variables con Permutation Importance
perm_importance = permutation_importance(log_reg_model, X_test, y_test_enc, scoring="accuracy", n_repeats=10, random_state=2025)

# Extraer las características y su importancia
importances = perm_importance.importances_mean
features = X_test.columns

# Ordenar los resultados
sorted_idx = np.argsort(importances)[::-1]

# Graficar  
plt.figure(figsize=(10,6))
plt.barh(features[sorted_idx], importances[sorted_idx], color="blue")
plt.xlabel("Importancia")
plt.ylabel("Características")
plt.title("Permutation Importance - Logistic Regression")
plt.show()
```

```python
# Gradient Boosting Classifier

# Definir el Pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('gradient_boosting', GradientBoostingClassifier(random_state=2025))
])
# Definir la cuadrícula de parámetros
param_grid = {
    'gradient_boosting__n_estimators': [50, 100, 150],
    'gradient_boosting__learning_rate': [0.01, 0.1, 0.2],
    'gradient_boosting__max_depth': [3, 5, 7]
}
# configurar GridSearchCV
grid_sear = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
# entrenar la busqueda de hiperparámetros
grid_sear.fit(X_train, y_train_enc)
# obtener el mejor modelo
best_model = grid_sear.best_estimator_
# evaluar el conjunto de prueba
y_pred_gb = best_model.predict(X_test)
accuracy= accuracy_score(y_test_enc, y_pred_gb)
f1 = f1_score(y_test_enc, y_pred_gb, average='weighted')
# mostrar resultados
print(f'mejores parametros: {grid_sear.best_params_}')
print(f"Gradient Boosting Accuracy: {accuracy:.2f}")
print(f"F1 Score: {f1:.2f}")
print("\n")

# Classification Report
print(classification_report(y_test_enc, y_pred_gb, target_names=label_encoder.classes_))
```

```python
# ROC Curve Gradient Boosting
 
RocCurveDisplay.from_estimator(best_model, X_test, y_test_enc)
plt.title("ROC Curve - Gradient Boosting")
plt.show()
```

```python
# Confusion Matrix Gradient Boosting

cm = confusion_matrix(y_test_enc, best_model.predict(X_test))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
disp.plot(cmap="Blues")
plt.title("Matriz de Confusión - Gradient Boosting")
plt.show()
```

```python
# Permutation Importance Gradient Boosting

# Calcular la importancia de las variables con Permutation Importance
perm_importance = permutation_importance(best_model, X_test, y_test_enc, scoring="accuracy", n_repeats=10, random_state=2025)

# Extraer las características y su importancia
importances = perm_importance.importances_mean
features = X_test.columns

# Ordenar los resultados
sorted_idx = np.argsort(importances)[::-1]

# Graficar 
plt.figure(figsize=(10,6))
plt.barh(features[sorted_idx], importances[sorted_idx], color="blue")
plt.xlabel("Importancia")
plt.ylabel("Características")
plt.title("Permutation Importance - Gradient Boosting")
plt.show()
```
