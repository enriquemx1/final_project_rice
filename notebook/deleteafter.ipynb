```python
import pandas as pd

# Cargar el dataset desde el archivo Excel
df = pd.read_excel('/workspaces/final_project_rice/data/Rice_Cammeo_Osmancik.xlsx')
# Características a analizar
features = ["Area", "Perimeter", "Major_Axis_Length", "Minor_Axis_Length", "Eccentricity", "Convex_Area", "Extent"]

# Calcular valores mínimos y máximos
limits = {feature: (df[feature].min(), df[feature].max()) for feature in features}

# Mostrar los resultados
for feature, (min_val, max_val) in limits.items():
    print(f"{feature}: Min={min_val}, Max={max_val}")

# mostrar las caracteristicas cuando class= Osmancik
osmancik_df = df[df['Class'] == 'Osmancik']
osmancik_features = osmancik_df[features]
print("\nCaracterísticas de Osmancik:")
print(osmancik_features)
# mostrar las caracteristicas cuando class= Cammeo
cammeo_df = df[df['Class'] == 'Cammeo']
cammeo_features = cammeo_df[features]
print("\nCaracterísticas de Cammeo:")
print(cammeo_features)
```
